### Canviar el format dels atributs a numérics
Dels atributs que queden podem observar que n'hi ha molts que són cadenes de texte, alguns són noms directament i altres són nombres que tenen algún afegit com '€', 'lbs', etc. Per tant, a continuació farem modificacions a aquells que són quasi numérics per que sí que ho siguin i amb aquells que són noms utilitzarem One-Hot Encoding.

Primer usarem la funció següent per passar els valors del cost del jugador, el seu salari i la seva "release clause" a float, és a dir, llevar les 'K' i 'M' i el símbol '€'

```python
def value_to_float(x):
    """
    From K and M to float.
    
    """
    x = x.replace('€', '')
    ret_val = 0.0
    
    if type(x) == float or type(x) == int:
        ret_val = x
    if 'K' in x:
        if len(x) > 1:
            ret_val = float(x.replace('K', ''))
        ret_val = ret_val *1000
    if 'M' in x:
        if len(x) > 1:
            ret_val = float(x.replace('M', ''))
        ret_val = ret_val * 1000000.0
    return ret_val
```

```python
df["Value"] = df["Value"].apply(value_to_float)
df["Wage"] = df["Wage"].apply(value_to_float)
df["Release Clause"] = df["Release Clause"].apply(value_to_float)
```

Ara eliminarem lbs del pes dels jugadors i canviarem l'altura de peus a centímetres amb la següent funció:

```python
def feet_to_cm(x):
    """
    From feet to cm.
    
    """
    s = x.split("'")
    ret_val = 0.0
    
    feet = float(s[0])
    inches = float(s[1])
    
    ret_val = feet*30.48
    ret_val = ret_val+(inches*2.54)
    
    return ret_val
```

```python
df["Weight"] = df["Weight"].replace({'lbs': ''}, regex=True)
df["Weight"] = pd.to_numeric(df["Weight"], downcast='float')
df["Height"] = df["Height"].apply(feet_to_cm)
```

Ara, fent ús de la funció suma() que hi ha a continuació, sumarem els valors de +2 i +3 que apareixen a les columnes de posició dels jugadors amb el nombre que hi ha a l'esquerra.

```python
def suma(x):
    
    if isinstance(x, int) :
        return x
    
    s = x.split("+")
    
    n1 = float(s[0])
    n2 = float(s[1])
    
    ret_val = n1+n2
    
    return ret_val
```

```python
df['LS'] = df['LS'].apply(suma)
df['ST'] = df['ST'].apply(suma)
df['RS'] = df['RS'].apply(suma)
df['LW'] = df['LW'].apply(suma)
df['LF'] = df['LF'].apply(suma)
df['CF'] = df['CF'].apply(suma)
df['RF'] = df['RF'].apply(suma)
df['RW'] = df['RW'].apply(suma)
df['LAM'] = df['LAM'].apply(suma)
df['CAM'] = df['CAM'].apply(suma)
df['RAM'] = df['RAM'].apply(suma)
df['LM'] = df['LM'].apply(suma)
df['LCM'] = df['LCM'].apply(suma)
df['CM'] = df['CM'].apply(suma)
df['RCM'] = df['RCM'].apply(suma)
df['RM'] = df['RM'].apply(suma)
df['LWB'] = df['LWB'].apply(suma)
df['LDM'] = df['LDM'].apply(suma)
df['CDM'] = df['CDM'].apply(suma)
df['RDM'] = df['RDM'].apply(suma)
df['RWB'] = df['RWB'].apply(suma)
df['LB'] = df['LB'].apply(suma)
df['LCB'] = df['LCB'].apply(suma)
df['CB'] = df['CB'].apply(suma)
df['RCB'] = df['RCB'].apply(suma)
df['RB'] = df['RB'].apply(suma)
df.head()
```

Aquesta taula és la que ens queda després d'haver fet aquestes operacions. Cal comentar que al jupyter notebook pujat al repositori, es mostren més taules després de cada pas per poder veure els canvis pas a pas però per comoditat de la lectura del pdf s'han eliminat.  

Finalment eliminarem les columnes que contenen noms com poden ser de clubs o països i crearem una columna per cada possible valor de l'atribut indicant amb un 1 o 0 si el jugador pertany a aquell país o club, és a dir, usarem One-Hot Encoding.

```python
nat = df.pop("Nationality")
df = pd.concat([df.reset_index(drop=True), pd.get_dummies(nat, prefix='nat').reset_index(drop=True)], axis=1, sort=False)

clb = df.pop("Club")
df = pd.concat([df.reset_index(drop=True), pd.get_dummies(clb, prefix='clb').reset_index(drop=True)], axis=1, sort=False)

pf = df.pop("Preferred Foot")
df = pd.concat([df.reset_index(drop=True), pd.get_dummies(pf, prefix='pf').reset_index(drop=True)], axis=1, sort=False)

wr = df.pop("Work Rate")
df = pd.concat([df.reset_index(drop=True), pd.get_dummies(wr, prefix='wr').reset_index(drop=True)], axis=1, sort=False)

bt = df.pop("Body Type")
df = pd.concat([df.reset_index(drop=True), pd.get_dummies(bt, prefix='bt').reset_index(drop=True)], axis=1, sort=False)

pos = df.pop("Position")
df = pd.concat([df.reset_index(drop=True), pd.get_dummies(pos, prefix='pos').reset_index(drop=True)], axis=1, sort=False)

pd.set_option('display.max_columns', 20)
df.head(5)
```

S'ha posat un máxim de columnes a 20 ja que degut a l'elevat nombre de columnes no es pot generar el PDF amb totes elles. A més com que es trunca la taula per la meitat podem veure com a la part del final, les darreres columnes es corresponen amb el One-Hot Encoding i hi ha un 1 en el cas que el jugador tengui aquell atribut.

## Predicció

### Organitzar sets
El primer que hem de fer es separar el data frame en un set d'entrenament i un set de testeig i llevam 'Value' del data frame ja que és el valor que volem predir.

```python
val = df.pop("Value")

X_train, X_test, y_train, y_test = train_test_split(df, val, test_size=0.33, random_state=42)

print("Tamany del data frame: "+str(len(df)))
print("Tamany del set d'entrenament: "+str(len(X_train)))
```

### Entrenar model
Ara entrenam el nostre model lienal amb les dades que hem separat específicament per l'entrenament.

```python
reg = linear_model.LinearRegression().fit(X_train, y_train)
```

### Prediccions
Una vegada entrenat el model obtenim les prediccions per al set de 'testeig'. És a dir, ja podem complir l'objectiu de la pràctica d'obtenir el valor d'un jugador sabent-ne els seus atributs.

```python
prediccions = reg.predict(X_test)
```

Abans de comparar les prediccions amb els valors reals, podem veure que y_test esta indexat amb els indexos del data frame (el primer té index 12491, el segon 7988, etc.) mentre que les prediccions comencen en 0 i acaben en 5912, per tant el que farem sera posar a y_test els indexos començant per 0
